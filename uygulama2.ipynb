## Matris çarpımı, ters alma, determinant, özdeğerler ve özvektörler.

```python
# İki matris tanımlayalım.
A = np.array([[2, 1], [1, 3]])
B = np.array([[1, 2], [3, 4]])
print(A)
print(B)
```

```python
# Matris çarpımı
C = np.dot(A, B)
print(C)
```

```python
# Matrisin tersini alma
A_inv = np.linalg.inv(A)
print(A_inv)
```

```python
# Matrisin transpozu (devriğini bulma)
A= np.array([[1, 2], [3, 4], [5, 6]])
A_T = A.T
print(A_T)
# ya da
A_T = np.transpose(A)
print(A_T)
```

```python
# Bir matrisin determinantını alma
A= np.array([[1, 2], [3, 4]])
try:
    det_A = np.linalg.det(A)
except np.linalg.LinAlgError:
    print("Bu matrisin determinantı alınamaz.")
else:
    print(det_A)
```

```python
# Özdeğerler ve özvektörler

eig_vals, eig_vecs = np.linalg.eig(A)
print(np.linalg.eig(A))
print("Özdeğerler: ", eig_vals)
print("Özvektörler: ", eig_vecs)
```

# Scipy ve Olasılık Dağılımları

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

# Normal dağılım (μ=0, σ=1)
x = np.linspace(-4, 4, 100)
pdf = stats.norm.pdf(x, loc=0, scale=1)  # Olasılık yoğunluk fonksiyonu
cdf = stats.norm.cdf(x, loc=0, scale=1)  # Birikimli dağılım fonksiyonu

plt.figure()
plt.plot(x, pdf, label="PDF")
plt.plot(x, cdf, label="CDF", linestyle="dashed")
plt.legend()
plt.title("Normal Dağılım")
plt.show()
```

```python
# Olasılık işlemleri
# Koşullu olasılık örneği: P(A|B) = P(B|A) * P(A) / P(B)

P_A = 0.1  # Olay A'nın olasılığı
P_B = 0.5  # Olay B'nin olasılığı
P_B_given_A = 0.8  # B'nin A koşulunda gerçekleşme olasılığı

P_A_given_B = (P_B_given_A * P_A) / P_B  # Bayes Teoremi

print("P(A|B):", P_A_given_B)
```

```python
# Basit bir fonksiyonun grafiği
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x)

plt.figure()
plt.plot(x, y, label="sin(x)", color='b')
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.title("Sinüs Fonksiyonu")
plt.legend()
plt.show()
```

# Türev ve İntegral Alma

```python
import sympy as sp

# Sembolik değişken tanımlama
x = sp.Symbol('x')

# Fonksiyon tanımlama
f = x**3 + 2*x**2 - 5*x + 3

# Türev alma
f_derivative = sp.diff(f, x)

# İntegral alma
f_integral = sp.integrate(f, x)

print("Fonksiyon:", f)
print("Türevi:", f_derivative)
print("İntegrali:", f_integral)
```

## Optimizasyon

```python
from scipy.optimize import minimize

# Fonksiyon tanımlama
def func(x):
    return x**4 - 3*x**3 + 2

# Başlangıç tahmini
x0 = 2

# Optimizasyon işlemi (minimum bulma)
result = minimize(func, x0)

print("Minimum Nokta:", result.x)
print("Minimum Değer:", result.fun)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# X aralığını belirle
x = np.linspace(-2, 3, 100)
y = x**4 - 3*x**3 + 2

# Minimum noktayı bulma
x0=0
result = minimize(func, x0)
min_x = result.x[0]
min_y = result.fun

# Grafiği çiz
plt.plot(x, y, label=r"$f(x) = x^4 - 3x^3 + 2$")
plt.scatter(min_x, min_y, color='red', label="Minimum Nokta", zorder=3)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(min_x, color='red', linestyle="dashed", linewidth=0.7)
plt.legend()
plt.title("Fonksiyonun Minimum Noktasını Bulma")
plt.show()
```

* Bu fonksiyonda minimum nokta aramak için başlangıç değerini 2 değil de 0 olarak verelim.
